### Using a combination of bs4 and selenium

```python
def amazon_scraper(producto):
    URL = f"https://www.amazon.com.mx/s?k={producto.replace(' ', '+')}"
    chrome_options = Options()
    chrome_options.add_argument('--headless') 
    driver=webdriver.Chrome(options=chrome_options)
    driver.get(URL)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    results = soup.find_all('div', {'data-component-type': 's-search-result'})
    productos=[]
    for i in range(len(results)):
        nombre=results[i].find_all("span", class_='a-size-base-plus a-color-base a-text-normal')[0].text
        try:
            puntuacion=results[i].find_all("span", class_='a-icon-alt')[0].text
        except:
            puntuacion='N/A'
        try:
            precio=results[i].find_all("span", class_='a-price-whole')[0].text
        except:
            precio='N/A'
        link='https://www.amazon.com.mx/'+results[i].find_all("a", class_='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal')[0]['href']
        productos.append({'Nombre':nombre,'Precio':precio,'Puntuacion':puntuacion,'Enlace':link})

    return pd.DataFrame(productos)
```

```python
def mercadolibre_scraper(producto):
    URL = f"https://listado.mercadolibre.com.mx/{producto.replace(' ', '+')}"
    chrome_options = Options()
    chrome_options.add_argument('--headless') 
    driver=webdriver.Chrome(options=chrome_options)
    driver.get(URL)
    soup = BeautifulSoup(driver.page_source, 'html.parser')    
    results = soup.find_all('div', {'class': 'ui-search-result__content'})
    if len(results)==0:
        results = soup.find_all('div', {'class': 'ui-search-result__content-wrapper'})

    productos=[]
    for i in range(len(results)):
        link=results[i].find_all('a', {'class': 'ui-search-item__group__element ui-search-link__title-card ui-search-link'})[0]['href']
        nombre=results[i].find_all('h2', {'class': 'ui-search-item__title'})[0].text
        try:
            puntuacion=results[i].find_all('span', {'class': 'andes-visually-hidden'})[0].text
        except:
            puntuacion='N/A'
        precio=results[i].find_all('span', {'class': 'andes-money-amount__fraction'})[0].text
        productos.append({'Nombre':nombre,'Precio':precio,'Puntuacion':puntuacion,'Enlace':link})

    return pd.DataFrame(productos)
```
